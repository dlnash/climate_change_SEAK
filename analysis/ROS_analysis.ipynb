```python
model = 'cfsr'
varname = 'snow'
# --- read the non-anomaly data ---
ds = load_preprocessed_WRF_data(model, varname, anomaly=False)

# --- calculate rain (PCPT - SWE) ---
ds['rain'] = ds['pcpt'] - ds['snow']

# --- update snow depth to mm ---
snowh_attrs = ds['snowh'].attrs
ds['snowh'] = ds['snowh'] * 1000
ds['snowh'].attrs = snowh_attrs
ds['snowh'].attrs['units'] = 'mm'

# --- calculate ROS intensity ---
# ROS intensity is defined as the change in snow depth (snowh) for the ROS day
ds['delsnow'] = ds['snow'].diff(dim='time') # n+1 - n
ds['delsnowh'] = ds['snowh'].diff(dim='time') # n+1 - n

# --- calculate ROS ---
# A ROS event is defined as a day when rain (pcpt) is greater than 5 mm d−1, 
# delta snowh (change in snow depth) decreases (< 0) and snow cover is true
# Create the mask and add it as a new variable called "ros"
ds['ros'] = ((ds['rain'] > 5) & (ds['delsnowh'] < 0) & (ds['snowc'] > 0)).astype(int)

# --- calculate ROS intensity --- 
# ROS intensity is the sum of rainfall (ds['rain']) and snowmelt (ds[delsnowh']*-1)
ds['ros_intensity'] = ds['rain'].where(ds['ros'] == 1) + (ds['delsnowh'].where(ds['ros'] == 1))*-1

# --- compute sum of rain-on-snow days per year ---
da_ros = ds['ros'].groupby("time.year").sum('time').rename(year="time")

# --- compute change in rainfall and snowmelt during ROS events
rain_mask = ds['rain'].where(ds['ros'] == 1)
# negative delsnowh indicates snow decreased between today and tomorrow
# need to multiply by 1
snowmelt_mask = (ds['delsnowh'].where(ds['ros'] == 1))*-1 

da_rain = rain_mask.groupby("time.year").mean('time').rename(year="time")
da_snowmelt = snowmelt_mask.groupby("time.year").mean('time').rename(year="time")
da_ros_intensity = ds['ros_intensity'].groupby("time.year").mean('time').rename(year="time")

# --- merge into single dataset ---
ds = xr.merge([da_ros, da_rain, da_snowmelt, da_ros_intensity])

ds
```

```python


# --- Sum over time: number of ROS days per grid cell ---
ros_sum = ds['ros'].sum(dim='time')

# Make sure we have lat/lon arrays
lons = ds['lon']
lats = ds['lat']

# --- Plot ---
fig = plt.figure(figsize=(8, 6))
ax = plt.axes(projection=ccrs.PlateCarree())

# Southeast Alaska extent
ax.set_extent([-142, -130, 54, 61], crs=ccrs.PlateCarree())

# Add geographic features
ax.add_feature(cfeature.COASTLINE, linewidth=0.8)
ax.add_feature(cfeature.BORDERS, linewidth=0.5)
ax.add_feature(cfeature.LAND, facecolor='lightgray', zorder=0)
ax.add_feature(cfeature.OCEAN, facecolor='white', zorder=0)
ax.gridlines(draw_labels=True, linestyle='--', color='gray', alpha=0.5)

# --- Contour plot ---
# Define contour levels for discrete ROS counts (0, 1, 2, 3)
levels = np.arange(1, 500, 25)  # finer levels between 0–3
cmap = plt.get_cmap('Blues')

cf = ax.contourf(
    lons, lats, ros_sum,
    levels=levels,
    cmap=cmap,
    vmin=0, vmax=500,
    extend='neither',
    transform=ccrs.PlateCarree()
)

# Add colorbar
cbar = plt.colorbar(cf, ax=ax, orientation='vertical', shrink=0.75, pad=0.02)
cbar.set_label('Number of ROS Days')

# Title and layout
plt.title('Rain-on-Snow (ROS) Days\nSoutheast Alaska', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
def find_nearest_indices(ds, lat, lon):
    # Function to find nearest grid indices on 2D curvilinear grid
    dist = (ds['lat'] - lat)**2 + (ds['lon'] - lon)**2
    iy, ix = np.unravel_index(dist.argmin(), dist.shape)
    return iy, ix

# --- select case ---
start_date = '2018-12-30'
end_date = '2019-01-04'
ds = ds.sel(time=slice(start_date, end_date))

# --- select location ---
lat_lst = [55.683099]
lon_lst = [-132.571308]
# Get indices for each requested point
indices = [find_nearest_indices(ds, la, lo) for la, lo in zip(lat_lst, lon_lst)]
iy = [i[0] for i in indices]
ix = [i[1] for i in indices]

# Subset the dataset
ds_point = ds.isel(
    y=xr.DataArray(iy, dims="location"),
    x=xr.DataArray(ix, dims="location")
)

ds_point
```

```python


# Example: assume ds is your xarray dataset with dims ('time', 'lat', 'lon')
# and variables: ivt, rain, snow, deltasnowh, ros

time = ds_point['time']

ivt = ds_point['ivt']
rain = ds_point['rain']
snow = ds_point['snow']
deltasnowh = ds_point['delsnowh']
ros = ds_point['ros']

# --- Start Plot ---
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot IVT (right axis, different scale)
ax2 = ax1.twinx()
ax2.plot(time, ivt, color='tab:purple', label='IVT (kg m$^{-1}$ s$^{-1}$)', linewidth=1.8)
ax2.set_ylabel('IVT (kg m$^{-1}$ s$^{-1}$)', color='tab:purple')
ax2.tick_params(axis='y', labelcolor='tab:purple')

# Plot precipitation variables on left axis
ax1.plot(time, rain, label='Rain (mm)', color='tab:blue', linewidth=1.5)
ax1.plot(time, snow, label='Snow (mm)', color='tab:cyan', linewidth=1.5)
ax1.plot(time, deltasnowh, label='ΔSnowH (mm)', color='tab:orange', linewidth=1.5)

ax1.set_xlabel('Time')
ax1.set_ylabel('Precipitation / Snow (mm)')
ax1.tick_params(axis='y', labelcolor='black')


# --- Shade where ROS == 1 (centered on each day) ---
ros_mask = ros.values.squeeze().astype(bool)

# Convert time coordinate to NumPy datetime64 for math operations
time_np = np.array(time.values)

# Width of shading (1 day centered on each timestamp)
half_day = np.timedelta64(12, 'h')

for t, r in zip(time_np, ros_mask):
    if r:
        ax1.axvspan(t - half_day, t + half_day,
                    color='lightgray', alpha=0.5, zorder=0)

# Add dashed 0 line
ax1.axhline(0, color='black', linestyle='--', linewidth=1)

# --- Final touches ---
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left', frameon=False)

plt.title('Time Series with Rain, Snow, ΔSnowH, IVT, and ROS Periods')
plt.tight_layout()
plt.show()
```
